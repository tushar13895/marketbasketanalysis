```python
# import required packages here
import numpy as np
import pandas as pd
import collections
from itertools import permutations
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# load the basket.csv file into a pandas DataFrame (df)
df = pd.read_csv('basket.csv')
# Use df.shape to see the number of rows and columns in the df
df.shape
```

```python
# Use df.head() to see the first 8 rows of the df
df.head(8)
```

```python
# Set the index of the df to the Transaction_ID column
df.set_index('ID', inplace=True)
# Use df.head() to see the first 3 rows of the df
df.head(3)
```

```python
# Use the pandas describe method to get an overview of the dataset
df.describe()
```

```python
# Replace nan entities with 'None' and assign to a new dataframe
df1 = df.astype(object).replace(np.nan, 'None')
```

```python
# create a list of lists, where each list contains the items in a given transaction
records = []
for i in range(0, len(df1)): 
    records.append([str(df1.values[i,j]) for j in range(0, len(df1.columns))])
# view the first transaction in the list
records[:1]
```

```python
# from your list of lists, create a flattened list that contains all items purchased
flattened = [item for record in records for item in record]
# view the legnth of the list
print(len(flattened))
```

```python
# remove any elements that items of interest (like empty strings or nan values)
flattened_new = [i for i in flattened if i != 'None']
# create a list of unique items from the flattened list
items = list(set(flattened_new))
# print out how many unique items we have
print('# of items:',len(items))
# print out the unique items
items.sort() 
print(items)
```

```python
# generate the itemset permutations up to 2-itemsets
rules = list(permutations(items, 2))
# print out the number of rules
print('# of rules:',len(rules))
# print some of the elements of the list of rules
print(rules[:5])
```

```python
# generate the frequency counts for each unique item in the dataset
item_freq = collections.Counter(flattened_new)
# show the most common
item_freq.most_common()
```

```python
# one-hot encode the data and show the first 5 rows of the resulting df
# don't forget to drop the 'nan' value

encoder = TransactionEncoder().fit(records)
onehot = encoder.transform(records)

df_onehot = pd.DataFrame(onehot, columns=encoder.columns_).drop('None', axis=1)
df_onehot
```

```python
# Generate frequent itemsets with a minimum support of 0.5%
two_itemsets = apriori(df_onehot, min_support=0.005, use_colnames=True)
# view the itemsets sorted by most frequent
two_itemsets.sort_values(by=['support'], ascending=False)
```

```python
# Generate frequent itemsets with a minimum support of 0.1%
two_itemsets1 = apriori(df_onehot, min_support=0.001, use_colnames=True)
# view the itemsets sorted by most frequent
two_itemsets1.sort_values(by=['support'], ascending=False)
```

```python
# Generate frequent itemsets with a minimum support of 1%
two_itemsets2 = apriori(df_onehot, min_support=0.01, use_colnames=True)
# view the itemsets sorted by most frequent
two_itemsets2.sort_values(by=['support'], ascending=False)
```

```python
# Generate frequent itemsets with a minimum support of 10%
two_itemsets3 = apriori(df_onehot, min_support=0.1, use_colnames=True)
# view the itemsets sorted by most frequent
two_itemsets3.sort_values(by=['support'], ascending=False)
```
